```python
from matplotlib import rc
%matplotlib inline
rc("font", family="serif", size=12)
rc("text", usetex=True)
import daft

pgm = daft.PGM([10, 10], origin=[0.7, 0])


# And a plate.
pgm.add_plate(daft.Plate([0.9, 2.6, 6, 4.5], label=r"known source $i$",
    shift=-0.1))
pgm.add_plate(daft.Plate([2.6, 0.5, 3, 3], label=r"pixel $j$",
    shift=-0.1))
pgm.add_plate(daft.Plate([2.5, 0.12, 4, 3.5], label=r"band $\lambda$",
    shift=-0.1))
```

```python
pgm.add_node(daft.Node("A", r"$A_{i,j,\lambda}$", 4, 3, fixed=True))
pgm.add_node(daft.Node("data", r"$D_{j,\lambda}$", 5, 1, observed=True))
pgm.add_node(daft.Node("model", r"$m_{j,\lambda}$", 5, 2))
pgm.add_node(daft.Node("background", r"$B_{\lambda}$", 6, 2))


pgm.add_node(daft.Node("f_ir", r"$f_{i,\lambda}$", 6, 3))
pgm.add_node(daft.Node("alpha", r"$\alpha_i$", 4, 4, fixed=True))
pgm.add_node(daft.Node("delta", r"$\delta_i$", 3, 4, fixed=True))
pgm.add_node(daft.Node("z", r"$z_i$", 2, 4, fixed=True))
pgm.add_node(daft.Node("r", r"$\mathbf{r}_i$", 3, 5, fixed=True))


```

```python
pgm.add_edge("model", "data")
pgm.add_edge("background", "model")
pgm.add_edge("A", "model")
pgm.add_edge("f_ir", "model")
pgm.add_edge("alpha", "A")
pgm.add_edge("delta", "A")
pgm.add_edge("r", "z",directed=False,plot_params={'ls':'dashed'})
pgm.add_edge("r", "alpha",directed=False,plot_params={'ls':'dashed'})
pgm.add_edge("r", "delta",directed=False,plot_params={'ls':'dashed'})

pgm.add_edge("r","f_ir")
```

```python
#high z stacking panel
pgm.add_plate(daft.Plate([7.5, 0.2, 3, 4.2], label=r"high z $\l$",
    shift=-0.1))
pgm.add_node(daft.Node("A_z", r"$A_{z,j,\lambda}$", 8.8, 3, fixed=True))
pgm.add_node(daft.Node("f_ir_z", r"$f_{l,\lambda}$", 8.8, 4))
pgm.add_node(daft.Node("alpha_z", r"$\alpha_l$", 9.8, 2, fixed=True))
pgm.add_node(daft.Node("delta_z", r"$\delta_l$", 8.8, 2, fixed=True))
pgm.add_node(daft.Node("z_z", r"$z_i$", 7.8, 2, fixed=True))
pgm.add_node(daft.Node("r_z", r"$\mathbf{r}_l$", 8.8, 1, fixed=True))
pgm.add_node(daft.Node("mu_z", r"$\mu$", 7.8, 5))
pgm.add_node(daft.Node("sig_z", r"$\sigma$", 8.8, 5))

pgm.add_edge("A_z", "model")
pgm.add_edge("f_ir_z", "model")
pgm.add_edge("alpha_z", "A_z")
pgm.add_edge("delta_z", "A_z")
pgm.add_edge("r_z", "z_z",directed=False,plot_params={'ls':'dashed'})
pgm.add_edge("r_z", "alpha_z",directed=False,plot_params={'ls':'dashed'})
pgm.add_edge("r_z", "delta_z",directed=False,plot_params={'ls':'dashed'})
pgm.add_edge("mu_z","f_ir_z")
pgm.add_edge("sig_z","f_ir_z")

```

```python
# Render and save.
pgm.render()
```

```python
pgm.figure.savefig("graphical_model_high_z.pdf")
```
