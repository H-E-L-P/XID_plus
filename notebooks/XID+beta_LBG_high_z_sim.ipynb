# XID+(beta) LBG high z simulation

In order to test whether XID+highz works, I am simulating the LBG sample by injecting sources into the GOODS-South map.
The sample will include 100 sources, whose flux will be normally distributed.

```python
import numpy as np
import pylab as plt
from astropy.io import fits
%matplotlib inline
import XIDp_mod_beta as xid_mod
from astropy import wcs
import pickle
import dill
```

```python
#Folder containing prior input catalogue
folder="/Users/pdh21/astrodata/goodss/"
#prior catalogue
prior_cat="cat_prior_24_goodss.fits"
hdulist = fits.open(folder+prior_cat)
fcat=hdulist[1].data
hdulist.close()
inra=fcat['RA_IRAC']
indec=fcat['DEC_IRAC']
```

```python
# Load up High Z sources for stacking

#Folder containing prior input catalogue
folder="/Users/pdh21/astrodata/high_z/"
#prior catalogue
prior_cat_stack="ZSOUTHDEEP.fits"
hdulist = fits.open(folder+prior_cat_stack)
fcat_z=hdulist[1].data
hdulist.close()
inra_z=fcat_z['RA']
indec_z=fcat_z['DEC']
nsrc_z=inra_z.size
```

## Create simulated sources

```python
mu_sim=20.0
sig_sim=3.0
nsrc_sim=100.0
np.random.seed(1)
flux_sim=(np.random.randn(nsrc_sim)*sig_sim)+mu_sim
```

```python
range_ra=np.max(inra_z)-np.min(inra_z)
range_dec=np.max(indec_z)-np.min(indec_z)

inra_sim=np.linspace(0.0,1.0,nsrc_sim**0.5)*range_ra+np.min(inra_z)
indec_sim=np.linspace(0.0,1.0,nsrc_sim**0.5)*range_dec+np.min(indec_z)
inra_sim,indec_sim=np.meshgrid(inra_sim,indec_sim)
inra_sim=inra_sim.flatten()
indec_sim=indec_sim.flatten()
```

The other 24 micron sources (red), the original stack sample in GOODS-South (Blue), and my simulated sources (Green)

```python
plt.scatter(inra,indec,c='r')
plt.scatter(inra_z,indec_z,c='b')
plt.scatter(inra_sim,indec_sim,c='g')
plt.xlabel('RA')
plt.ylabel('DEC')
```

# Prepare data and simulated data for fit

```python
#Folder containing maps
#imfolder='/research/astrodata/fir/hermes/xid/XID_2014/SMAP_images_v4.2/' 
imfolder='/Users/pdh21/astrodata/SMAP_images_v4.2/'
#field
field='GOODS-S'
#SMAP version
SMAPv='4.2'


# In[3]:

pswfits=imfolder+field+'_image_250_SMAP_v'+SMAPv+'.fits'#SPIRE 250 map
pmwfits=imfolder+field+'_image_350_SMAP_v'+SMAPv+'.fits'#SPIRE 350 map
plwfits=imfolder+field+'_image_500_SMAP_v'+SMAPv+'.fits'#SPIRE 500 map
```

```python
#-----250-------------
hdulist = fits.open(pswfits)
im250phdu=hdulist[0]
im250=hdulist[1].data*1.0E3
nim250=hdulist[2].data*1.0E3
w_250 = wcs.WCS(hdulist[1].header)
pixsize250=3600.0*w_250.wcs.cd[1,1] #pixel size (in arcseconds)
hdulist.close()
#-----350-------------
hdulist = fits.open(pmwfits)
im350phdu=hdulist[0]
im350=hdulist[1].data*1.0E3
nim350=hdulist[2].data*1.0E3
w_350 = wcs.WCS(hdulist[1].header)
pixsize350=3600.0*w_350.wcs.cd[1,1] #pixel size (in arcseconds)
hdulist.close()
#-----500-------------
hdulist = fits.open(plwfits)
im500phdu=hdulist[0]
im500=hdulist[1].data*1.0E3
nim500=hdulist[2].data*1.0E3
w_500 = wcs.WCS(hdulist[1].header)
pixsize500=3600.0*w_500.wcs.cd[1,1] #pixel size (in arcseconds)
hdulist.close()
```

```python
#pixsize array (size of pixels in arcseconds)
pixsize=np.array([pixsize250,pixsize350,pixsize500])
#point response function for the three bands
prfsize=np.array([18.15,25.15,36.3])
#set fwhm of prfs in terms of pixels
pfwhm=prfsize/pixsize
#set size of prf array (in pixels)
paxis=[13,13]
#use Gaussian2DKernel to create prf (requires stddev rather than fwhm hence pfwhm/2.355)
from astropy.convolution import Gaussian2DKernel

prf250=Gaussian2DKernel(pfwhm[0]/2.355,x_size=paxis[0],y_size=paxis[1])
prf250.normalize(mode='peak')
prf350=Gaussian2DKernel(pfwhm[1]/2.355,x_size=paxis[0],y_size=paxis[1])
prf350.normalize(mode='peak')
prf500=Gaussian2DKernel(pfwhm[2]/2.355,x_size=paxis[0],y_size=paxis[1])
prf500.normalize(mode='peak')
```

```python
prior250=xid_mod.prior(prf250,im250,nim250,w_250,im250phdu)
prior250.prior_bkg(-4.0,2.0)
prior250.prior_cat(inra,indec,prior_cat)
prior250.prior_cat_stack(inra_sim,indec_sim,'highz_sim')
prior250.get_pointing_matrix()
prior250.get_pointing_matrix_coo()
```

```python
from scipy.sparse import coo_matrix
fvec=np.empty((prior250.nsrc+1))
fvec[:]=0.0
fvec[-1-prior250.stack_nsrc:-1]=flux_sim
fvec[-1]=0.0
f=coo_matrix((fvec, (range(0,prior250.nsrc+1),np.zeros(prior250.nsrc+1))), shape=(prior250.nsrc+1, 1))
```

```python
x_range=np.max(prior250.sx_pix)-np.min(prior250.sx_pix)
y_range=np.max(prior250.sy_pix)-np.min(prior250.sy_pix)
rmap=(prior250.A*f)
im_temp=np.empty_like(im250)
im_temp[prior250.sy_pix,prior250.sx_pix]=np.asarray(rmap.todense()).reshape(-1)
```
