```python
def plot_fit(prior,fvec):
    from scipy.sparse import coo_matrix
    

    x_range=np.max(prior.sx_pix)-np.min(prior.sx_pix)
    y_range=np.max(prior.sy_pix)-np.min(prior.sy_pix)
    f=coo_matrix((fvec, (range(0,prior.nsrc+1),np.zeros(prior.nsrc+1))), shape=(prior.nsrc+1, 1))
    A=coo_matrix((prior.amat_data, (prior.amat_row, prior.amat_col)), shape=(prior.snpix, prior.nsrc+1))
    rmap_temp=(A*f)
    im_temp=np.empty((x_range+1,y_range+1))
    
    im_temp[prior.sx_pix-np.min(prior.sx_pix),prior.sy_pix-np.min(prior.sy_pix)]=np.asarray(rmap_temp.todense()).reshape(-1)#+np.random.randn(prior.snpix)*prior.snim]
    return im_temp

def plot_fit_orig(prior):
    from scipy.sparse import coo_matrix
    
    x_range=np.max(prior.sx_pix)-np.min(prior.sx_pix)
    y_range=np.max(prior.sy_pix)-np.min(prior.sy_pix)
    f=coo_matrix((fvec, (range(0,prior.nsrc+1),np.zeros(prior.nsrc+1))), shape=(prior.nsrc+1, 1))
    A=coo_matrix((prior.amat_data, (prior.amat_row, prior.amat_col)), shape=(prior.snpix, prior.nsrc+1))
    rmap_temp=(A*f)
    im_temp=np.empty((x_range+1,y_range+1))
    
    im_temp[prior.sx_pix-np.min(prior.sx_pix),prior.sy_pix-np.min(prior.sy_pix)]=prior.sim
    return im_temp
def plot_fit_orig_nim(prior):
    from scipy.sparse import coo_matrix
    
    x_range=np.max(prior.sx_pix)-np.min(prior.sx_pix)
    y_range=np.max(prior.sy_pix)-np.min(prior.sy_pix)
    f=coo_matrix((fvec, (range(0,prior.nsrc+1),np.zeros(prior.nsrc+1))), shape=(prior.nsrc+1, 1))
    A=coo_matrix((prior.amat_data, (prior.amat_row, prior.amat_col)), shape=(prior.snpix, prior.nsrc+1))
    rmap_temp=(A*f)
    im_temp=np.empty((x_range+1,y_range+1))
    
    im_temp[prior.sx_pix-np.min(prior.sx_pix),prior.sy_pix-np.min(prior.sy_pix)]=prior.snim
    return im_temp
```

```python
fvec=np.empty(([obj['psw'].nsrc+1]))
fvec[0:-1]=XIDdata['flux250']
fvec[-1]=XIDdata['bkg250'][0]
print fvec
```

```python
im_temp=plot_fit(obj['psw'],fvec)
plt.imshow(im_temp)
plt.colorbar()
```

```python
im_temp_orig=plot_fit_orig(obj['psw'])
im_temp_orig_nim=plot_fit_orig_nim(obj['psw'])

plt.imshow(im_temp_orig_nim)
plt.colorbar()

```
